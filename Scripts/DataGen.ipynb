```python
import pandas as pd
import numpy as np

# Create a DataFrame for each table
years = range(2002, 2024)  # 5 years from 2020 to 2025
districts =["Nyarugenge", "Gasabo", "Kicukiro",'Nyagatare','Gatsibo','Kayonza','Rwamagana',
            'Bugesera','Ngoma','Kirehe','Rulindo','Gakenke','Musanze','Burera','Gicumbi','Kamonyi',
            'Muhanga','Ruhango','Nyanza','Huye','Nyamagabe','Gisagara','Nyaruguru',
            'Nyamasheke','Rusizi','Karongi','Ngororero','Rutsiro','Rubavu','Nyabihu']

# Blood Smears Model
blood_smears_data = {
    "Year": np.repeat(years, len(districts)),
    "District": np.tile(districts, len(years)),
    "BloodSmearsCollected": np.random.randint(1000, 5000, size=len(years) * len(districts)),
    "BloodSmearsExamined": np.random.randint(800, 4000, size=len(years) * len(districts)),
    "PositiveTest": np.random.randint(350, 2500, size=len(years) * len(districts)),
    "NegativeTest": np.random.randint(600, 2000, size=len(years) * len(districts)),


}
blood_smears_df = pd.DataFrame(blood_smears_data)

# Malaria Incidence Model
malaria_incidence_data = {
    "Year": np.repeat(years, len(districts)),
    "District": np.tile(districts, len(years)),
    "AnnualParasiteIncidence": np.random.uniform(5, 50, size=len(years) * len(districts)),
    "SlidePositivityRate": np.random.uniform(1, 10, size=len(years) * len(districts)),
    "AnnualBloodExaminationRate": np.random.uniform(20, 80, size=len(years) * len(districts)),
    "AnnualVivaxIncidence": np.random.uniform(2, 20, size=len(years) * len(districts)),
    "AnnualMalariaeIncidence": np.random.uniform(2, 20, size=len(years) * len(districts)),
    "AnnualOvaleIncidence": np.random.uniform(2, 20, size=len(years) * len(districts)),
    "AnnualFalciparumIncidence": np.random.uniform(2, 20, size=len(years) * len(districts)),



}
malaria_incidence_df = pd.DataFrame(malaria_incidence_data)

# Malaria Cases Model
malaria_cases_data = {
    "Year": np.repeat(years, len(districts)),
    "District": np.tile(districts, len(years)),
    "TotalCases": np.random.randint(100, 1000, size=len(years) * len(districts)),
    "PvCases": np.random.randint(20, 200, size=len(years) * len(districts)),
    "PfCases": np.random.randint(80, 800, size=len(years) * len(districts)),
    "PmCases": np.random.randint(10, 50, size=len(years) * len(districts)),
    "PoCases": np.random.randint(35, 150, size=len(years) * len(districts)),

    "MixedCases": np.random.randint(5, 50, size=len(years) * len(districts)),
    }
malaria_cases_df = pd.DataFrame(malaria_cases_data)

# Malaria Treatment and Mortality Model
malaria_treatment_data = {
    "Year": np.repeat(years, len(districts)),
    "District": np.tile(districts, len(years)),
    "RadicalTreatment": np.random.randint(50, 500, size=len(years) * len(districts)),
    "TotalDeaths": np.random.randint(10, 100, size=len(years) * len(districts)),
    "PvDeaths": np.random.randint(2, 20, size=len(years) * len(districts)),
    "PfDeaths": np.random.randint(8, 80, size=len(years) * len(districts)),
    "PmDeaths": np.random.randint(1, 10, size=len(years) * len(districts)),
    "PoDeaths": np.random.randint(5, 35, size=len(years) * len(districts)),
}
malaria_treatment_df = pd.DataFrame(malaria_treatment_data)
```
